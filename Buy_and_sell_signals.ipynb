```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
 
from pandas_datareader import data
%matplotlib inline
start = '2020-03-01'
end = '2021-03-31'
df = data.DataReader('^N225', 'yahoo', start, end)
df.head(10)
```

```python
date = df.index
price = df['Adj Close']
```

```python
plt.plot(date, price)
```

```python
plt.figure(figsize = (20,5))
plt.plot(date, price, label = 'Nikke225')
plt.title('N225', color = 'blue', backgroundcolor = 'white', size = 12, loc = 'center')
plt.legend()
```

```python
plt.figure(figsize = (20,5))
plt.plot(date, price, label = 'Nikke225')
plt.title('N225', color = 'blue', backgroundcolor = 'white', size = 15, loc = 'center')
plt.xlabel('日付', color = 'blue', size = 15)
plt.ylabel('価格', color = 'blue', size = 15)

plt.legend()
```

```python
sma25 = 25
sma50 = 50

df['sma25'] = price.rolling(window = sma25).mean()
df['sma50'] = price.rolling(window = sma50).mean()
```

```python
pd.set_option('display.max_rows', None)
df.head(100)
```

```python
plt.figure(figsize = (20,5))

plt.plot(date, price, label = 'Nikke225')

plt.plot(date, df['sma25'], label = '25日移動平均')
plt.plot(date, df['sma50'], label = '50日移動平均')

plt.title('N225', color = 'blue', backgroundcolor = 'white', size = 15, loc = 'center')
plt.xlabel('日付', color = 'blue', size = 15)
plt.ylabel('価格', color = 'blue', size = 15)

plt.legend()
```
